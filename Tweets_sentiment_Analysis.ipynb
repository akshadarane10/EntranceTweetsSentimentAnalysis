# single frequency counter by passing the text file

```python
import re
 
frequency = {}
document_text = open('keywords.txt', 'r')
text_string = document_text.read().lower()
match_pattern = re.findall(r'\b[a-z]{3,15}\b', text_string)
 
for word in match_pattern:
    count = frequency.get(word,0)
    frequency[word] = count + 1
     
frequency_list = frequency.keys()
 
for words in frequency_list:
    print(words, frequency[words])
```

# counting number of repeated words in each row/each Tweets(Function)

```python
#re.split(r',|-|:'

def word_count(str, refdict):
    counts = dict()
    words = str.replace('@', ' ').split()
    

    for word in words:
        if word in refdict:
            if word in counts:
                counts[word] += 1
            else:
                counts[word] = 1
    return counts
```

# Sentiment Analysis

```python
from textblob import TextBlob
```

```python
def getSubjectivity(text1):
    return TextBlob(text1).sentiment.subjectivity


def getPolarity(text1):
    return TextBlob(text1).sentiment.polarity

df['Subjectivity'] = df['Tweet_clean'].apply(getSubjectivity)
df['Polarity'] = df['Tweet_clean'].apply(getPolarity)

df.head(50)
```

```python
def getAnalysis(score):
    if score<0:
        return 'Negative'
    elif score==0:
        return 'Neutral'
    else:
        return 'Positive'
    
df['Analysis'] = df['Polarity'].apply(getAnalysis)

df.head()
```
